```python
import cv2
import os

# read video from file
cam = cv2.VideoCapture('./Dataset/train/Basketball/v_Basketball_g01_c01.avi')
```

```python
video_fps = cam.get(cv2.CAP_PROP_FPS),
total_frames = cam.get(cv2.CAP_PROP_FRAME_COUNT)
height = cam.get(cv2.CAP_PROP_FRAME_HEIGHT)
width = cam.get(cv2.CAP_PROP_FRAME_WIDTH)

print(f"Frame Per second: {video_fps } \nTotal Frames: {total_frames} \n Height: {height} \nWidth: {width}")
```

```python
# set the directory containing training videos
videos_dir = './Dataset/train/Basketball'

# set the directory to store extracted frames
frames_dir = './data'

# creating a folder named frames_dir
if not os.path.exists(frames_dir):
    os.makedirs(frames_dir)

# loop over each video in the directory
for video_file in os.listdir(videos_dir):
    if not video_file.endswith('.avi'):
        continue
    # read the video file
    cam = cv2.VideoCapture(os.path.join(videos_dir, video_file))
    currentframe = 0

    # loop over each frame in the video
    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = os.path.join(frames_dir, f'{video_file[:-4]}_frame{currentframe}.jpg')
            print(f'Creating...{name}')

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

    # Release the video capture and move to the next video
    cam.release()

# Release all space and windows once done
cv2.destroyAllWindows()
```

```python
# Python program to create
# Image Classifier using CNN

# Importing the required libraries
import numpy as np
from random import shuffle
from tqdm import tqdm

'''Setting up the env'''

TRAIN_DIR = './data'
TEST_DIR = './data2'
IMG_SIZE = 100
LR = 1e-3


'''Setting up the model which will help with tensorflow models'''
MODEL_NAME = 'sports-{}-{}.model'.format(LR, '6conv-basic')

'''Labelling the dataset'''
def label_img(img):
    word_label = img.split('.')[0]
    if word_label == 'Basketball': return [1, 0, 0]
    elif word_label == 'BenchPress': return [0, 1, 0]
    elif word_label == 'BaseballPitch': return [0, 0, 1]

'''Creating the training data'''
def create_train_data():
    # Creating an empty list where we should store the training data
    # after a little preprocessing of the data
    training_data = []

    # tqdm is only used for interactive loading
    # loading the training data
    for img in tqdm(os.listdir(TRAIN_DIR)):

        # labeling the images
        label = label_img(img)

        path = os.path.join(TRAIN_DIR, img)

        # loading the image from the path and then converting them into
        # grayscale for easier covnet prob
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

        # resizing the image for processing them in the covnet
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

        # final step-forming the training data list with numpy array of the images
        training_data.append([np.array(img), np.array(label)])

    # shuffling of the training data to preserve the random state of our data
    shuffle(training_data)

    # saving our trained data for further uses if required
    np.save('train_data.npy', training_data)
    return training_data

'''Processing the given test data'''
# Almost same as processing the training data but
# we dont have to label it.
def process_test_data():
    testing_data = []
    for img in tqdm(os.listdir(TEST_DIR)):
        path = os.path.join(TEST_DIR, img)
        img_num = img.split('.')[0]
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        testing_data.append([np.array(img), img_num])

    shuffle(testing_data)
    np.save('test_data.npy', testing_data)
    return testing_data

'''Running the training and the testing in the dataset for our model'''
train_data = create_train_data()
test_data = process_test_data()

# train_data = np.load('train_data.npy')
# test_data = np.load('test_data.npy')
'''Creating the neural network using tensorflow'''
# Importing the required libraries
import tflearn
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression

import tensorflow as tf
tf.compat.v1.reset_default_graph()
convnet = input_data(shape =[None, IMG_SIZE, IMG_SIZE, 1], name ='input')

convnet = conv_2d(convnet, 32, 5, activation ='relu')
convnet = max_pool_2d(convnet, 5)
 
convnet = conv_2d(convnet, 64, 5, activation ='relu')
convnet = max_pool_2d(convnet, 5)
 
convnet = conv_2d(convnet, 128, 5, activation ='relu')
convnet = max_pool_2d(convnet, 5)
 
convnet = conv_2d(convnet, 64, 5, activation ='relu')
convnet = max_pool_2d(convnet, 5)
 
convnet = conv_2d(convnet, 32, 5, activation ='relu')
convnet = max_pool_2d(convnet, 5)
 
convnet = fully_connected(convnet, 1024, activation ='relu')
convnet = dropout(convnet, 0.8)
 
convnet = fully_connected(convnet, 3, activation ='softmax')
convnet = regression(convnet, optimizer ='adam', learning_rate = LR,
      loss ='categorical_crossentropy', name ='targets')

model = tflearn.DNN(convnet, tensorboard_dir ='log') 

X = np.array([i[0] for i in train_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
Y = []
for i in train_data:
    if i[1] == 0:
        Y.append([1, 0, 0])
    elif i[1] == 1:
        Y.append([0, 1, 0])
    else:
        Y.append([0, 0, 1])
Y = np.array(Y)

test_x = np.array([i[0] for i in test_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
test_y = []
for i in test_data:
    if i[1] == 0:
        test_y.append([1, 0, 0])
    elif i[1] == 1:
        test_y.append([0, 1, 0])
    else:
        test_y.append([0, 0, 1])
test_y = np.array(test_y)

model.fit({'input': X}, {'targets': Y}, n_epoch=5,
          validation_set=({'input': test_x}, {'targets': test_y}),
          snapshot_step=500, show_metric=True, run_id=MODEL_NAME)
```

# Problem of overfitting:

The original Image size was 50 and epochs 10. The results were accurate but the model was overfitting and taking too much noise. Therefore the Image size was increased to 300 to allow the model to train better however with the same number of epochs the traineing time was greatly increased.In our third attempt the number of epochs were reduced and so was the Image size (IMG = 100, epochs = 5) to get a better result and to solve the overfitting problem 

```python
#Testing data
import numpy as np
import matplotlib.pyplot as plt
# if you need to create the data:
# test_data = process_test_data()
# if you already have some saved:
test_data = np.load('test_data.npy', allow_pickle = True)
 
fig = plt.figure()
 
for num, data in enumerate(test_data[:20]):
    img_num = data[1]
    img_data = data[0]
     
    y = fig.add_subplot(4, 5, num + 1)
    orig = img_data
    data = img_data.reshape(IMG_SIZE, IMG_SIZE, 1)
 
    model_out = model.predict([data])[0]
     
    if np.argmax(model_out) == 0:
        str_label ='Basketball'
    elif np.argmax(model_out) == 1:
        str_label ='BenchPress'
    else:
        str_label ='BaseballPitch'
         
    y.imshow(orig, cmap ='gray')
    plt.title(str_label)
    y.axes.get_xaxis().set_visible(False)
    y.axes.get_yaxis().set_visible(False)
plt.show()
```

```python
# Use the trained model to predict the classes of test data
predictions = model.predict(test_x)
# Convert predicted values from one-hot encoding to class labels
predicted_classes = np.argmax(predictions, axis=1)
# Convert true values from one-hot encoding to class labels
true_classes = np.argmax(test_y, axis=1)
# Calculate the accuracy by comparing predicted and true classes
accuracy = np.mean(predicted_classes == true_classes)
print("Accuracy:", accuracy)
```
